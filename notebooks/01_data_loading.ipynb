# Data Loading

```python
import pandas as pd
import os
import glob
```

```python
RAW_DATA_PATH = '../data/raw/'
csv_files = glob.glob(os.path.join(RAW_DATA_PATH, '*.csv'))

dfs = []
for file in csv_files:
    df = pd.read_csv(file)
    dfs.append(df)

motor_data = pd.concat(dfs, ignore_index=True)
motor_data = motor_data.sort_values('timestamp').reset_index(drop=True)

print(f'Total records: {len(motor_data)}')
motor_data.head()
```
