# Sequence Builder

```python
import pandas as pd
import numpy as np
```

```python
def create_sequences(data, seq_length=30):
    X, y = [], []
    for i in range(len(data) - seq_length):
        X.append(data[i:i+seq_length, :-1])
        y.append(data[i+seq_length, -1])
    return np.array(X), np.array(y)

motor_data = pd.read_csv('../data/processed/processed_merged_sensorData.csv')
features = ['Air temperature [K]', 'Process temperature [K]', 'Rotational speed [rpm]', 'Torque [Nm]', 'Tool wear [min]', 'Machine failure']
data = motor_data[features].values

X, y = create_sequences(data, seq_length=30)

print(f'X shape: {X.shape}')
print(f'y shape: {y.shape}')
print(f'Failure sequences: {int(y.sum())} / {len(y)} ({y.mean():.2%})')

np.save('../data/processed/X_sequences.npy', X)
np.save('../data/processed/y_sequences.npy', y)
```
